```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver = web_driver()

driver.get('https://www.google.com')

driver.quit()
```

```python
from selenium.webdriver.common.by import By
```

```python
job_titles = []
job_listings = []

for job in job_listings:
    try:
        # Access the span with class 'sr-only' within the job listing
        job_title_element = job.find_element(By.TAG_NAME, "span", class_="sr-only")
        job_title = job_title_element.text.strip()
        job_titles.append(job_title)
    except NoSuchElementException:
        # Handle cases where the 'sr-only' span might not be present
        print("Job title element not found for a listing. Skipping.")

# Print or process the scraped job titles
print("Scraped Job Titles:")
for title in job_titles:
    print(title)
```

```python
import csv
from google.colab import files

job_description_to_search = "data scientist"

# Prepare the CSV file
csv_file_path = 'job_listings.csv'

with open('job_listings.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Job Title', 'Address', 'Job Link'])  # Write the header row

    for listing in j_l:
        try:

            # Get the job title
            j_e = listing.find_element(By.CLASS_NAME, 'sr-only')
            j_t = j_e.text

            # Get the address
            a_e = listing.find_element(By.CLASS_NAME, 'base-search-card__metadata')
            address = a_e.text

            # Get the job link
            job_link_element = listing.find_element(By.TAG_NAME, 'a')
            job_link = job_link_element.get_attribute('href')

            # Write the row to the CSV file
            writer.writerow([j_t, address, job_link])

            # Print the output for debugging
            print("Job Title:", j_t)
            print("Address:", address)
            print("Job Link:", job_link)
            print("-" * 50)

        except NoSuchElementException:
            print("Element not found for a listing. Skipping.")
```
